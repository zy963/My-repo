```python
# Import AutoML things
import numpy as np

from pycaret.regression import *
```

```python
# Set up the model
settings = setup(data = dataset, 
          target = 'body_mass_g', 
          session_id = 123
          ) 
```

```python
# Train a bunch of different models and save the top 3
top3 = compare_models(n_select = 3)
compare_model_results = pull()
```

```python
# Stack the top 3 models
stacker_top3 = stack_models(top3)
```

```python
# First, save a model
save_model(stacker_top3, "best_model")
```

```python
asdf = load_model("best_model")
```

```python
def predict(species, island, culmen_length_mm, culmen_depth_mm, flipper_length_mm, sex):
  # species	island	culmen_length_mm	culmen_depth_mm	flipper_length_mm	sex
    df = pd.DataFrame.from_dict({'species': [species], 'island': [island], 
                    'culmen_length_mm': [culmen_length_mm], 'culmen_depth_mm': [culmen_depth_mm],
                    'flipper_length_mm': [flipper_length_mm], 'sex': [sex]})
    
    pred = predict_model(asdf, df)
    return pred['Label'][0].astype('float64')
```

```python
predict('Adelie', 'Torgersen',	39.1,	18.7,	181.0, 'MALE')
```
